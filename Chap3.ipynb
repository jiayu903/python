## List[串列]
使用中括號[ ]包裹元素

```python
score=[90,100,88,60,79]
print(score[0]) #第1個資料
print(score[1])  #第2個資料

print(score[-1]) #倒數第1個資料
print(score[-2])  #倒數第2個資料
```

```python
#score=[90,100,88,60,79]
for i in range(len(score)): #len(score): score的長度
  print(score[i])
```

```python
#score=[90,100,88,60,79] --> score[0],score[1],score[2],score[3],score[4]
print(score[0:3]) #取出score[0]到score[2],不包含score[3]
print(score[:]) #取出所有值
print(score[:4]) #取出score[0]到score[3],不包含score[4]
print(score[2:]) #取出score[2]後面的所有值
print(score[0:5:2]) #從score[0]:不包含score[5]:間隔2
print(score[::2])  #間隔2

print(score[::-2]) #間隔-2:代表從列表最後往前取值,間隔2
print(score[4:0:-2])  #從score[4]:不包含score[0]:間隔2 -->[79,60,88,100] --->[79,88]
```

list的長度


```python
my_list0=[] #空串列
print(my_list0)
my_list1=list() #空串列
my_list2=['python','javascript','c#'] #my_list2=[my_list2[0],my_list2[1],my_list2[2]]
print(my_list2)
print(len(my_list2)) #len:計算list的長度
print(my_list2[2])
```

append附加, remove移除, insert插入, clear清除


```python
my_list2=['python','javascript','c#']
my_list2.append('java')   #append:附加在最後面
print(my_list2)
my_list2.remove('java')  #remove:移除
print(my_list2)
my_list2.insert(0,'java') #insert:插入物件，需要指定索引位置
print(my_list2)
my_list2.clear()  #clear:清除所有物件
print(my_list2)
```

list相加&次數

```python
a=[1,2,3]
b=[4,5,6]
c=a+b  #list串列相加
print(c)
d=a*2  #list串列次數
print(d)
```
